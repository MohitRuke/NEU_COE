## Writing your Blog

**Cleary describe one major idea**

What exactly is your topic?  What are you trying to do? It should be clear what you are doing within the first couple paragraphs.  If not, stop and clearly explain what you are doing?

Use a real world example, if need be. Examples are particularly useful for subjects the audience may not be familaiar.  


**Give people a reason to care**

Stir your audience’s curiosity, by hinting or telling them what they'll learn.  

**Mix writing and graphics with code**

Don't write too much without the audience actually doing something by programming.  Six hundred words or so is a good rule of thumb.  If you get to six hundred words; consider running some code.  

**Know your audience**

Know you audience, which should be Masters students in data science. Don't spend too much time explaining concepts they should know.  You can always link to other resources to refresh their memory.  

* Identify the audience.  
* Research and structure the content.  
* Ensure it is accurate and understood.  

**Readable code**

Write readable code. Use a style guide like [Google Python Style Guide](https://google.github.io/styleguide/pyguide.html).  

**Cleary define how you are evaluting your model**

Cleary define how you are evaluting your model. Vague terms like accuracy, and/or cross validation are not enough.  Typically an equation that describes how you are calculating something like accuracy is required.  

**Try out your draft on a volunteer listener**

* Did I explain my idea clearly?  
* Was there anything in my talk that you didn’t follow?  
* Was there anything you didn’t understand?  
* Did anything seem out of place?  
* Did I lose your interest anywhere?  

**Make sure your audience can run your code**

* Can the code run on different computers?
* Be clear about any dependencies and/pr requirements.  

**Include links to more information at the end**

Include links to more information at the end. 


**Include a blurb about youself at the end**

If you would like people to contact you, include a blurb about youself at the end as well as a valid e-mail.
